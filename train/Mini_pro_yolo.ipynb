```python
import pandas as pd
import matplotlib.pyplot as plt

# โหลดข้อมูลการเทรนจากไฟล์ CSV
df = pd.read_csv("/content/runs/detect/train/results.csv")

# แสดงค่าทั้งหมด
print(df.head())

# วาดกราฟ Loss
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train/box_loss'], label="Train Box Loss", color='blue')
plt.plot(df['epoch'], df['val/box_loss'], label="Val Box Loss", color='red')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()

# วาดกราฟ Accuracy (mAP)
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label="Train mAP", color='green')
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.legend()
plt.title("Training Accuracy (mAP)")
plt.show()
```

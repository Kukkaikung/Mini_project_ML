```python
import pandas as pd
import matplotlib.pyplot as plt

# à¹‚à¸«à¸¥à¸”à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸à¸²à¸£à¹€à¸—à¸£à¸™à¸ˆà¸²à¸à¹„à¸Ÿà¸¥à¹Œ CSV
df = pd.read_csv("/content/runs/detect/train/results.csv")

# à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸²à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
print(df.head())

# à¸§à¸²à¸”à¸à¸£à¸²à¸Ÿ Loss
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train/box_loss'], label="Train Box Loss", color='blue')
plt.plot(df['epoch'], df['val/box_loss'], label="Val Box Loss", color='red')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()

# à¸§à¸²à¸”à¸à¸£à¸²à¸Ÿ Accuracy (mAP)
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label="Train mAP", color='green')
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.legend()
plt.title("Training Accuracy (mAP)")
plt.show()
```
